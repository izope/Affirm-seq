```r
library(Seurat)
library(dplyr)
library(ggplot2)
library(patchwork)

root <- "/Users/fupac/Desktop/DECODER单细胞/figure2/Affirmseq_EpCAM"
obj <- readRDS(file.path(root, "Affirmseq_EpCAM.rds"))
fd <- obj@misc$figure_data
```

```r
roi <- fd$roi %>% mutate(cell = factor(cell), px = as.numeric(cell), py = 1)
p_l <- ggplot(roi, aes(px, py)) + geom_rect(aes(xmin = 0.35, xmax = 5.65, ymin = 0.45, ymax = 1.55), inherit.aes = FALSE, fill = NA, color = "#E53935", linewidth = 0.7) + geom_point(aes(fill = EPCAM_expr), shape = 21, size = 13, color = "white", stroke = 1.5) + geom_text(aes(label = cell), y = 0.65, color = "white", size = 5, family = "Arial") + annotate("text", x = 3, y = 1.82, label = "RNA sequencing", color = "white", size = 6, family = "Arial") + annotate("segment", x = 0.1, xend = 1.25, y = 0.18, yend = 0.18, arrow = arrow(length = unit(0.16, "inches")), color = "white") + annotate("segment", x = 0.1, xend = 0.1, y = 0.18, yend = 0.98, arrow = arrow(length = unit(0.16, "inches")), color = "white") + annotate("text", x = 1.35, y = 0.11, label = "X", color = "white", size = 6, family = "Arial") + annotate("text", x = 0.02, y = 1.08, label = "Y", color = "white", size = 6, family = "Arial") + scale_fill_gradient(low = "#100B07", high = "#F5A623", name = "Gene\nexpression") + coord_cartesian(xlim = c(-0.05, 6.4), ylim = c(0.05, 2.05), expand = FALSE) + theme_void() + theme(plot.background = element_rect(fill = "black", color = NA), panel.background = element_rect(fill = "black", color = NA), legend.position = "right", legend.title = element_text(color = "white", size = 14), legend.text = element_text(color = "white"))
ggsave(file.path(root, "Affirmseq_EpCAM_panel_l_rna_sequencing.pdf"), p_l, width = 7, height = 3, device = cairo_pdf)
p_l
```

```r
bar <- fd$bar %>% mutate(cell = factor(cell), type = factor(type, levels = c("Protein", "Gene")))
p_m <- ggplot(bar, aes(cell, value, fill = type)) + geom_col(position = position_dodge(width = 0.75), width = 0.7) + scale_fill_manual(values = c(Protein = "#377EB8", Gene = "#E41A1C"), name = NULL) + scale_y_continuous(limits = c(0, 1.1), expand = c(0, 0)) + labs(x = "Cell", y = "Normalized value") + theme_classic(base_size = 16) + theme(text = element_text(family = "Arial", color = "black"), axis.text = element_text(color = "black"), legend.position = c(0.82, 0.83))
ggsave(file.path(root, "Affirmseq_EpCAM_panel_m_bar.pdf"), p_m, width = 4.2, height = 3.2, device = cairo_pdf)
p_m
```

```r
corr <- fd$corr
p_n <- ggplot(corr, aes(protein_norm, gene_norm)) + geom_smooth(method = "lm", se = FALSE, color = "black", linewidth = 0.9) + geom_point(size = 3.5, color = "#377EB8") + annotate("text", x = 0.02, y = 1.03, label = paste0("Pearson r = ", sprintf("%.2f", fd$pearson)), hjust = 0, size = 5.2, family = "Arial") + scale_x_continuous(limits = c(-0.02, 1.08), expand = c(0, 0)) + scale_y_continuous(limits = c(-0.02, 1.08), expand = c(0, 0)) + labs(x = "EpCAM protein (normalized)", y = "EPCAM gene (normalized)") + theme_classic(base_size = 16) + theme(text = element_text(family = "Arial", color = "black"), axis.text = element_text(color = "black"))
ggsave(file.path(root, "Affirmseq_EpCAM_panel_n_correlation.pdf"), p_n, width = 4.2, height = 3.7, device = cairo_pdf)
p_n
```
